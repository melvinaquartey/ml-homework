```python
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2lab
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from sklearn.cluster import KMeans
```

```python
image = imread('C:\\Users\\melvi\\Downloads\\seg2.jpg')
```

```python
lab_image = rgb2lab(image)
```

```python
reshaped_image = lab_image.reshape((-1, 3))
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(reshaped_image)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
clustered_image = cluster_labels.reshape(image.shape[:2])
plt.imshow(clustered_image, cmap='viridis')
plt.axis('off')
plt.show()
```

### I noticed that one different runs, clustering performs differently on the various objects in the image


```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from sklearn.metrics import silhouette_score 

silhouette_scores = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k,init='k-means++',
    n_init='auto',)
    kmeans.fit(reshaped_image)
    silhouette_scores.append(silhouette_score(reshaped_image, kmeans.labels_))

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), silhouette_scores)
plt.grid(True)
plt.xticks(ticks = range(2, 20))
plt.title('silhouette_scores')
plt.xlabel('k')
plt.ylabel('silhouette scores')
```

### It appears that cluster numbers between 2-5 (with silhouette scores greater than 0.5) are optimal for implementing in the k-means clustering

```python
kmeans = KMeans(n_clusters=2, init='k-means++', n_init='auto',)
kmeans.fit(reshaped_image)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
clustered_image = cluster_labels.reshape(image.shape[:2])
plt.imshow(clustered_image, cmap='viridis')
plt.axis('off')
plt.show()
```
