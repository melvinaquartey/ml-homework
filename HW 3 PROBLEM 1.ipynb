```python
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2lab
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from sklearn.cluster import KMeans
```

```python
image = imread('C:\\Users\\melvi\\Downloads\\seg2.jpg')
```

```python
lab_image = rgb2lab(image)
```

```python
reshaped_image = lab_image.reshape((-1, 3))
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(reshaped_image)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
clustered_image = cluster_labels.reshape(image.shape[:2])
plt.imshow(clustered_image, cmap='viridis')
plt.axis('off')
plt.show()
```

### I noticed that one different runs, clustering performs differently on the various objects in the image
### So I went on to experiment with different numbers of clusters and different initialization methods for clustering (kmeans++ or random) 

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from skimage import io, color
import numpy as np

# Load the image

# Convert the image to a 2D array
reshaped_image = np.reshape(image, (-1, 3))

# Experiment with different values for n_clusters and init
n_clusters_values = [3, 5, 7]
init_values = ['k-means++', 'random']

# Plot the segmented images for each combination of hyperparameters
fig, axes = plt.subplots(len(n_clusters_values), len(init_values), figsize=(15, 10))

for i, n_clusters in enumerate(n_clusters_values):
    for j, init in enumerate(init_values):
        # Perform KMeans clustering
        kmeans = KMeans(n_clusters=n_clusters, init=init)
        kmeans.fit(reshaped_image)
        clustered_image = kmeans.labels_.reshape(image.shape[:2])

        # Plot the segmented image
        ax = axes[i, j]
        ax.imshow(clustered_image, cmap='viridis')
        ax.set_title(f'n_clusters={n_clusters}, init={init}')
        ax.axis('off')

plt.tight_layout()
plt.show()
```
